```python
data = {"Salary":[3500,3800,4100,4400,4700,5000,5200,5250]*2}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.plot(df["Salary"],color="red",marker="o")
```

```python
plt.boxplot(df["Salary"])
```

```python
df["dept"]=['HR','IT','Finance','QA','Dev','HR','IT','Data Analyst']*2
```

```python
count=df["dept"].value_counts()
```

```python
# Custom explode mapping per department
explode_map = {
    "HR": 0.05,
    "IT": 0.1,
    "Finance": 0.2,
    "QA": 0,
    "Dev": 0.15,
    "Data Analyst": 0.25
}

# Build explode list in the same order as count.index
explode = [explode_map[dept] for dept in count.index]
```

```python
plt.pie(count,labels=count.index,explode=explode, autopct="%1.1f")
```

```python
plt.bar(count.index, count)
```
