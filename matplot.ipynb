```python
data = {"Salary":[3500,3800,4100,4400,4700,5000,5200,5250]*2}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.plot(df["Salary"],color="red",marker="o",linestyle=":",linewidth="3")
```

```python
plt.hist(df["Salary"],bins=5,color="green")
```

```python
plt.boxplot(df["Salary"])
```

```python
df["dept"]=['HR','IT','Finance','QA','Finance','HR','IT','Data Analyst']*2
```

```python
df.head()
```

```python
count = df["dept"].value_counts()
```

```python
count
```

```python
explode_map = {
    "HR": 0.05, "IT": 0.1, "Finance": 0.2, "QA": 0, "Dev": 0.15, "Data Analyst": 0.25}

explode = [explode_map[dept] for dept in count.index]
```

```python
plt.pie(count,labels=count.index,autopct="%1.1f", explode=explode)
```

```python
plt.bar(count.index,count)
```

```python
df["Age"]=[22,23,24,25,26,30,32,33,34,45,46,48,50,54,57,59]
```

```python
df
```

```python
plt.scatter(df["Age"],df["Salary"],color="orange")
```

```python
plt.plot(df["Age"],df["Salary"],marker="*")
plt.grid()
```

```python
plt.bar(df["Age"],df["Salary"])
```

```python
df
```

```python
hr_sal = df[df["dept"]=="HR"]["Salary"]
it_sal = df[df["dept"]=="IT"]["Salary"]
fin_sal = df[df["dept"]=="Finance"]["Salary"]
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal],labels=["HR","IT","Fin"])
```

```python
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(["HR","IT","Fin"],[hr_mean,it_mean,fin_mean])
```
