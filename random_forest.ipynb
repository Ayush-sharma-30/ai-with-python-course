*This code trains a Random Forest model on the handwritten digits dataset, predicts digits*

```python
import pandas as pd
from sklearn.datasets import load_digits
# contains images of handwritten digits (0–9).
digits = load_digits()
```

```python
# Shows all attributes available inside the digits object (like data, images, target, etc).
# digits.data → flattened pixel values (features)
# digits.images → actual 8×8 images of digits
# digits.target → labels (the actual digit: 0–9)

dir(digits)
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.gray() sets default colormap to grayscale (black/white) so digit images look natural.
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
df = pd.DataFrame(digits.data)
# each row = one image, each column = one pixel (64 total).
df.head()
```

```python
df['target'] = digits.target
```

```python
df[0:12]
```

```python
X = df.drop('target',axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
# n_estimators play with this to fine tune model
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
